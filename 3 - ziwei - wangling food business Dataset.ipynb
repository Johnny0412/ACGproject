### Tidy up 1: Rename all the variables

```r
colnames(businesses_mpi) <- c("site_registration_number", "registration_authority", "verification_agency", "registration_tpye", "fcp_template_type", "rbm/site_status",
                 "expiry_date", "legal_name", "trading_name", "physical_location", "city", "sectors", "product_types", "trading_operations", "processes_of_interest")

businesses_mpi %>%
    head()
```

### Tidy up 2: Missing data - "FCP Template Type" - "N/A";  "Processes of Interest" - "Nil"

```r
businesses_mpi = businesses_mpi %>%
    mutate(fcp_template_type  = na_if(fcp_template_type, "N/A")) %>%
    mutate(processes_of_interest  = na_if(processes_of_interest, "Nil")) 
    
businesses_mpi %>%
  skim()
```

### Tidy up 3: Convert column "Expiry Date" to Date type

```r
businesses_mpi = businesses_mpi %>%
    mutate(expiry_date  = as.Date(expiry_date, "%d/%m/%Y"))

businesses_mpi %>%
  head()
```

### Tidy up 4: Extract Suburbs out of "Physical Locaion"

```r
businesses_mpi %>%
  head(6)
```

```r
businesses_mpi = businesses_mpi %>%
    mutate(suburb = sub(".*, ", "", physical_location)) %>%
    mutate(suburb = na_if(suburb, "[Private]")) %>%
    mutate(suburb = case_when(city == "Akaroa" ~ "Akaroa",
                              city == "Lyttelton" ~ "Lyttelton",
                               TRUE ~ suburb)) 
           
businesses_mpi %>%
  head(6)
```

### Tidy up 5: "Trading Operation" separate the options out as individual flagged columns; maybe do the same for "Sector", "Product Types", "Processes of interest"

```r
trading_operations = businesses_mpi %>%
            select("trading_operations")
```

```r
trading_operations
```

```r
trading_operations %>%
    mutate(unpacked_trading_operations = str_split(trading_operations, ";")) %>%
    unnest %>% 
    mutate(n = 1)  %>%
    select(unpacked_trading_operations, n) %>%
    group_by(unpacked_trading_operations) %>%
    tally() %>%
    arrange(desc(n))
    
```

The aboved checks the unique trading operations in the unpacked list.
<br>The interesting info I would llike to extract for people using this data is 
<br>-Takeaway
<br>-Eat in premises
<br>-Retail
<br>-Caterer
<br>-Home delivery
<br>-Wholesale

```r
businesses_mpi = businesses_mpi %>%
    mutate(takeaway_flag = case_when(str_detect(trading_operations, "Takeaway") ~ 1,TRUE ~ 0)) %>%
    mutate(eat_in_premises_flag = case_when(str_detect(trading_operations, "Eat in premises") ~ 1,TRUE ~ 0)) %>%
    mutate(retail_flag = case_when(str_detect(trading_operations, "Retail") ~ 1,TRUE ~ 0)) %>%
    mutate(caterer_flag = case_when(str_detect(trading_operations, "Caterer") ~ 1,TRUE ~ 0)) %>%
    mutate(home_delivery_flag = case_when(str_detect(trading_operations, "Home delivery") ~ 1,TRUE ~ 0)) %>%
    mutate(wholesale_flag = case_when(str_detect(trading_operations, "Wholesale") ~ 1,TRUE ~ 0)) 
           
businesses_mpi %>%
  head(15)
```

```r
product_types = businesses_mpi %>%
            select("product_types")
```

```r
product_types
```

```r
product_types %>%
    mutate(unpacked_product_types = str_split(product_types, ";")) %>%
    unnest %>% 
    mutate(n = 1)  %>%
     select(unpacked_product_types, n) %>%
     group_by(unpacked_product_types) %>%
    tally() %>%
    arrange(desc(n))
```

The aboved checks the unique Product Types in the unpacked list
<br>The interesting info I would llike to extract for people using this data (number of the count >= 50 looked at) is 
<br>-Ready-to-eat meals & snacks
<br>-Shelf-stable products
<br>-Baked products
<br>-Eggs
<br>-fruits & vegetables 
<br>-Processed meat, poultry & seafood products
<br>-Sushi
<br>-Hot beverages

```r
businesses_mpi = businesses_mpi %>%
    mutate(ready_to_eat_meals_snacks_flag = case_when(str_detect(product_types, "Ready-to-eat meals & snacks") ~ 1,TRUE ~ 0)) %>%
    mutate(shelf_stable_food_flag = case_when(str_detect(product_types, "Shelf-stable") ~ 1,TRUE ~ 0)) %>%
    mutate(baked_food_flag = case_when(str_detect(product_types, "Baked") ~ 1,TRUE ~ 0)) %>%
    mutate(eggs_flag = case_when(str_detect(product_types, "Eggs") ~ 1,TRUE ~ 0)) %>%
    mutate(fruit_veg_flag = case_when(str_detect(product_types, "fruits & vegetables") ~ 1,TRUE ~ 0)) %>%
    mutate(meat_poultry_seafood_flag = case_when(str_detect(product_types, "meat, poultry") ~ 1,TRUE ~ 0)) %>%
    mutate(sushi_flag = case_when(str_detect(product_types, "Sushi") ~ 1,TRUE ~ 0)) %>%
    mutate(hot_beverages_flag = case_when(str_detect(product_types, "Hot beverages") ~ 1,TRUE ~ 0)) %>%
    mutate(any_business_count = 1)
           
businesses_mpi %>%
  head(6)
```

Next, build a summary table per suburb and counting the number of flags and total number of businesses.
In the checking of suburb level, I found that Hornby" is listed as a separate suburb, so I will correct that manually so it gets merged as Hornby suburb.

```r
businesses_mpi = businesses_mpi %>%
    mutate(suburb = case_when(suburb == "Hornby\"" ~ "Hornby",
                              suburb == "North Linwood" ~ "Linwood",
                              suburb == "Redwood Park" ~ "Redwood",
                              suburb == "Riccarton Park" ~ "Riccarton",
                              suburb == "Chaneys" ~ "Belfast",
                               TRUE ~ suburb)) 
    
```

```r
suburb_level_food_business_summary = businesses_mpi %>%
        filter(!is.na(suburb)) %>%
        select(suburb, 
               any_business_count,
               takeaway_flag,
               eat_in_premises_flag,
               retail_flag,
               caterer_flag,
               home_delivery_flag,
               wholesale_flag,
               ready_to_eat_meals_snacks_flag,
               shelf_stable_food_flag,
               baked_food_flag,
               eggs_flag,
               fruit_veg_flag,
               meat_poultry_seafood_flag,
               sushi_flag,
               hot_beverages_flag
              ) %>%
        group_by(suburb) %>%
        summarise(cnt_total_business = sum(any_business_count), 
                  cnt_takeaway = sum(takeaway_flag),
                  cnt_eat_in_premises = sum(eat_in_premises_flag),
                  cnt_retail = sum(retail_flag),
                  cnt_caterer = sum(caterer_flag),
                  cnt_home_delivery = sum(home_delivery_flag),
                  cnt_wholesale = sum(wholesale_flag),
                  cnt_ready_to_eat_meals_snacks = sum(ready_to_eat_meals_snacks_flag),
                  cnt_shelf_stable_food = sum(shelf_stable_food_flag),
                  cnt_baked_food = sum(baked_food_flag),
                  cnt_eggs = sum(eggs_flag),
                  cnt_fruit_veg = sum(fruit_veg_flag),
                  cnt_meat_poultry_seafood = sum(meat_poultry_seafood_flag),
                  cnt_sushi = sum(sushi_flag),
                  cnt_hot_beverages = sum(hot_beverages_flag)               
                 ) %>%
       arrange(desc(cnt_total_business))

suburb_level_food_business_summary
```

```r
suburb_level_food_business_summary %>%
  write_csv("suburb_level_food_business_summary.csv")
```
